```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# Generate sample data
x = np.arange(10)  # Sample x values
y_line = x**2  # Sample y values for the line plot
y_bar = np.random.randint(0, 10, 10)  # Sample y values for the bar plot
y_scatter = np.random.randint(0, 10, 10)  # Sample y values for the scatter plot

# Create line plot
plt.figure(figsize=(12, 4))
plt.subplot(1, 4, 1)
plt.plot(x, y_line)
plt.title('Line Plot')

# Create bar plot
plt.subplot(1, 4, 2)
plt.bar(x, y_bar)
plt.title('Bar Plot')

# Create scatter plot
plt.subplot(1, 4, 3)
plt.scatter(x, y_scatter)
plt.title('Scatter Plot')

# Create correlation heatmap
# Generate some random correlation data
correlation_data = np.random.rand(10, 10)

# Create a DataFrame
df = pd.DataFrame(correlation_data, columns=[f'Feature {i}' for i in range(1, 11)])

plt.subplot(1, 4, 4)
plt.imshow(df.corr(), cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(df.columns)), df.columns, rotation=45)
plt.yticks(range(len(df.columns)), df.columns)
plt.title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```
