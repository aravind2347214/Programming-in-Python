```python

'''Multilevel Inheritance'''

class Client:
    def __init__(self,cid,name):
        self.cid=cid
        self.name=name

class Account(Client):
    def __init__(self, cid, name,accNo,accType):
        super().__init__(cid, name)
        self.accNo=accNo
        self.accType=accType

class Payment(Account):
    def __init__(self, cid, name, accNo, accType,pid,amount):
        super().__init__(cid, name, accNo, accType)
        self.pid=pid
        self.amount=amount

    def display(self):
        print("Client ID     : ",self.cid)
        print("Client Name   : ",self.name)   
        print("Account Number: ",self.accNo)   
        print("Account Type  : ",self.accType)   
        print("Payment ID    : ",self.pid)  
        print("Payment Amount: ",self.amount)      

cid=input("Enter Client ID")
name=input("Enter Client Name")
accNo=input("Enter Account Number")
accType=input("Enter Account Type")
pid=input("Enter Payment ID")
amount=input("Enter Payment Amount")
pObj=Payment(cid,name,accNo,accType,pid,amount)
pObj.display()


```

```python
'''Hierarchical  Inheritance'''

class User:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender

class Worker(User):
    def __init__(self, name, age, gender,workerId,workType):
        super().__init__(name, age, gender)
        self.workerId=workerId
        self.workType=workType

    def WorkerDisplay(self):
        print("Worker ID    : ",self.workerId)
        print("Worker Name  : ",self.name)   
        print("Worker Age   : ",self.age)   
        print("Worker Gender: ",self.gender)   
        print("Work Type    : ",self.workType)  

        

class Client(User):
    def __init__(self, name, age, gender,clientId,clientLoc):
        super().__init__(name, age, gender)
        self.clientId=clientId
        self.clientLoc=clientLoc

    def ClientDisplay(self):
        print("Client ID      : ",self.clientId)
        print("Client Name    : ",self.name)   
        print("Client Age     : ",self.age)   
        print("Client Gender  : ",self.gender)   
        print("Client Location: ",self.clientLoc)  


print("Enter Client Details")
cid=input("Enter Client ID")
cName=input("Enter Client Name")
cAge=input("Enter Client Age")
cGender=input("Enter Client Gender")
cLoc=input("Enter Client Location")
clientObj=Client(cid,cName,cAge,cGender,cLoc)


print("Enter Worker Details")
wid=input("Enter Worker ID")
wName=input("Enter Worker Name")
wAge=input("Enter Worker Age")
wGender=input("Enter Worker Gender")
wWorkType=input("Enter Work Type")
workerObj=Worker(wid,wName,wAge,wGender,wWorkType)

clientObj.ClientDisplay()
workerObj.WorkerDisplay()
```

```python
'''Multiple inheritance'''
class User:
    def __init__(self,name,age,gender):
        self.name=name
        self.age=age
        self.gender=gender

class Work:
    def __init__(self,workID,workLocation):
        self.workID =  workID
        self.workLocation=workLocation

class Worker(User,Work):
    def __init__(self, name, age, gender,workID,workLocation,workerID,workerRate):
        User.__init__(self,name, age, gender) 
        Work.__init__(self,workID,workLocation) 
        self.workerID=workerID
        self.workerRate=workerRate

    def WorkerDisplay(self):
        print("Worker ID     : ",self.workerID)
        print("Worker Name   : ",self.name)   
        print("Worker Age    : ",self.age)   
        print("Worker Gender : ",self.gender)   
        print("Work ID       : ",self.workID)
        print("Work Location : ",self.workLocation)
        print("Worker Rate   : ",self.workerRate)

print("Enter Worker Details")
workerID=input("Enter Worker ID")
wName=input("Enter Worker Name")
wAge=input("Enter Worker Age")
wGender=input("Enter Worker Gender")
workID=input("Enter Work ID")
workLocation=input("Enter Work Location")
workerRate=input("Enter Worker Hourly Rate")

workerObj=Worker(wName,wAge,wGender,workID,workLocation,workerID,workerRate)
workerObj.WorkerDisplay()


    
        

        
```
