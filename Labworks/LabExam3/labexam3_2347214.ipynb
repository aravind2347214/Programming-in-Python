```python
import pandas as pd
df= pd.read_csv('UScereal.csv')
```

```python
print("Summary and information About Dataset")
print(df.describe())
print(df.info())
if(df.isnull):
    print("\nThere are missing values")

```

```python
"""Average Protien Value of each Manufacutre """

averageProtienPerManufacutrer = df.groupby('mfr')['protein'].mean()
print(averageProtienPerManufacutrer)
```

```python
"""NAme of the cereal with high sugar from manufactrer"""
MaxSugarContent = df[df['mfr'] == 'G']['sugars'].max()
MaxSugarCereal = df[(df['mfr'] == 'G') & (df['sugars'] == MaxSugarContent)]['Name'].iloc[0]
print(MaxSugarCereal)
```

```python
"""
Generate any three different suitable plots using matplotlib library for the following. Use appropriate formatting like title, legend, x-value, y-value ect. (3x4=12) 
a. Using only shelf or vitamins 
b. Using mfr and fiber 
c. Using sodium and sugars 
"""
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df['shelf'].value_counts().plot(kind='bar',color=["red","blue","green"]).legend(['shelf-3','self-1',"shelf-2"])
plt.title('Distribution of cereals according to  Shelf')
plt.xlabel('shelf')
plt.ylabel('No: of cereals')
plt.show()


plt.figure(figsize=(10, 6))
df.groupby('mfr')['fibre'].mean().plot(kind='bar')
plt.title('Average Fiber content by manufacturer')
plt.xlabel('manufacturer name')
plt.ylabel('Average fiber content')
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(df['sodium'], df['sugars'], alpha=1)
plt.title('Scatter Plot between sodium and sugars')
plt.xlabel('sodium')
plt.ylabel('sugars')
plt.grid(True)
plt.show()
```
